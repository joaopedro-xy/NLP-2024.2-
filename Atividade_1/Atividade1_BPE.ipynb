Atividade 1- Implementando BPE 

Observação : Decidi treinar em cima somente dos 100 primeiros textos para dar tempo de entregar a atividade no prazo,
uma vez que o treinamento em cima de 10,000 textos consume muito tempo,
porém o codigo funciona perfeitamente para qualquer tamanho.

```python
from bpe import *
```

```python
caminho_corpus = 'corpus'

# Carregando os textos
texto_treino = carregar_textos(caminho_corpus)
tamanho_vocabulario = 300 # hiperparametro com valor a ser decidido


# apenas instanciando substituições e vocabulário
substituicoes, vocab = tokenizer_setup(texto_treino, tamanho_vocabulario)
```

**Mostrando o primeiro texto do corpus que servirá como nossa string de teste para tokenização**

```python
texto_teste= texto_treino[0]
texto_teste
```

**Gerando seus tokens com a  função `decode_ids_to_strings`**

```python
texto_codificado = encode_string(texto_teste, substituicoes, vocab)
strings_decodificadas = decode_ids_to_strings(texto_codificado, vocab)
print("Tokens:", strings_decodificadas)
```

**Sua codificação em inteiro de utf- 8**

```python
print("Texto codificado em ints-utf8 dentro do vocabulário:", texto_codificado)
```

**Decodificando os inteiros utf-8 para string original**

```python
texto_decodificado = decode(texto_codificado, vocab)
print("Texto decodificado:", texto_decodificado)
```
